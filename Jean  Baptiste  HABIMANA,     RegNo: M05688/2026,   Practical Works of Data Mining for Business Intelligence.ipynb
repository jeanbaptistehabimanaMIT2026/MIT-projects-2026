** MY TURN: NumPy Practice**

1. Create the unit_prices array

```python
import numpy as np
```

```python

unit_prices = np.array([450, 1200, 3000, 750, 2100, 980])
print("Unit prices:", unit_prices)
```

 2. Print the total (sum)

```python
print("Total sum:", np.sum(unit_prices))
```

 3. Print the average (mean), rounded

```python
print("Average price (rounded):", round(np.mean(unit_prices)))
```

4. Apply 5% price increase

```python
increased_prices = unit_prices * 1.05
print("Increased prices:", increased_prices)
```

 5. Boolean mask for prices > 1000

```python
mask = unit_prices > 1000
print("Prices greater than 1000:", unit_prices[mask])
```
